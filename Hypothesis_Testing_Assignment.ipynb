Problem 1

### A F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units. A randomly selected sample of cutlets was collected from both units and measured? Analyze the data and draw inferences at 5% significance level. Please state the assumptions and tests that you carried out to check validity of the assumptions.


### Import libraries

```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

### Reading the csv file

```python
cutlets=pd.read_csv("Cutlets.csv")
cutlets.head()
```

### H0 - There is no significant difference in the diamaters of unit A and unit B (diameteres of the cutlets are same)
### H1 - There is a significant difference in the diamaters of unit A and unit B (diameteres of the cutlets are not same)

### Performing a two tail T Test, to check the hypothesis we made

```python
stats.ttest_ind(cutlets["Unit A"],cutlets["Unit B"])
```

Creating Histogram


```python
plt.hist(cutlets['Unit A'])
```

```python
plt.hist(cutlets['Unit B'])
```

Creating a Scatterplot

```python
plt.plot(cutlets['Unit A'],cutlets['Unit B'],'bo')
```

Creating a Distribution Plot

```python
sns.distplot(cutlets['Unit A'])
```

```python
sns.distplot(cutlets['Unit B'])
```

## Conclusion:
### Pvalue = 0.47>0.05. So, we accept H0.
### There is no significant difference in the diamaters of unit A and unit B (diameteres of the cutlets are same)

Problem 2

A hospital wants to determine whether there is any difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list. They collected a random sample and recorded TAT for reports of 4 laboratories. TAT is defined as sample collected to report dispatch.
   
Analyze the data and determine whether there is any difference in average TAT among the different laboratories at 5% significance level.




```python
import pandas as pd
from scipy import stats
```

```python
labtat=pd.read_csv("LabTAT.csv")
labtat.head()
```

H0 - There is no difference in average of TAT value among the different laboratories

H1 - There is a difference in average of TAT value among the different laboratories

```python
stats.f_oneway(labtat.iloc[:,0], labtat.iloc[:,1],labtat.iloc[:,2],labtat.iloc[:,3])
```
